# Data Pre-processing

```python
# Removing puntuations, stopwords and lemmatizing the data

def text_process(text):
    # text = text.translate(str.maketrans('', '', string.punctuation))
    # text = [word for word in text.split() if word.lower() not in stop_words]
    text = [lemmatizer.lemmatize(word.lower()) \
            for word in word_tokenize(re.sub(r'([^\s\w]|_)+', ' ', str(text))) \
            if word.lower() not in stop_words]
    return " ".join(text)
```

```python
newsdata_df['text'] = newsdata_df['text'].apply(text_process)
```

## Converting words to vectors using TFIDF Vectorizer

```python
tfidf_model = TfidfVectorizer(max_features=200)

tfidf_df = pd.DataFrame(tfidf_model.fit_transform(newsdata_df['text']).todense())
tfidf_df.columns = sorted(tfidf_model.vocabulary_)
```

```python
tfidf_df.head(10)
```

# K-Means clusting using elbow function

```python
distortions = []
K = range(1,6)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(tfidf_df)
    distortions.append(sum(np.min(cdist(tfidf_df, kmeanModel.cluster_centers_, 'euclidean'), \
                                  axis=1)) / tfidf_df.shape[0])

plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal number of clusters')
plt.show()
```

```python
wcss_list = [] # intializing the list for the values of WCSS

# using the for loop for iterations form 1 to 6

for i in range(1, 6):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(tfidf_df)
    wcss_list.append(kmeans.inertia_)

plt.plot(range(1, 6), wcss_list)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of Clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(tfidf_df)
y_kmeans = kmeans.predict(tfidf_df)
newsdata_df['obtained_clusters'] = y_kmeans
```

```python
pd.crosstab(newsdata_df['category'].replace({0:'sci.electronics', 1:'misc.forsale', 2:'talk.religion.misc'}),\
            newsdata_df['obtained_clusters'].replace({0 : 'cluster_1', 1 : 'cluster_2', 2 : 'cluster_3', 3: 'cluster_4'}))
```
